```python
#basic linear plotting:
import matplotlib.pyplot as plt
xaxis=[25,26,27,28,29,30]
y1_axis=[28000,29000,30000,31000,32000,33000]
plt.plot(xaxis,y1_axis,color='k',linestyle='--',marker='o',linewidth=2,label='y1_axis')
y2_axis=[28000,32000,49000,50000,65000,75000]
plt.plot(xaxis,y2_axis,color='#5a7d9a',marker='d',label='y2_axis')
y3_axis=[38900,45000,62678,78900,85000,90000]
plt.plot(xaxis,y3_axis,color='y',linestyle='-.',marker='s',linewidth=3,label='y2_axis')
plt.style.use('fivethirtyeight')
plt.xlabel('ages')
plt.ylabel('salary')
plt.title('basic linear plot')
plt.tight_layout()
plt.legend()
plt.grid(False)
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

# subplots

```python
#figures and subplots
x=np.linspace(1,10,100)
y=np.log(x)
plt.figure(1)
plt.subplot()
plt.title('y=log(x)')
plt.plot(x,y)

plt.figure(2)
plt.subplot()
plt.title('y=log(x)**2')
plt.plot(x,y**2)

plt.show()
```

```python
x=np.linspace(1,10,100)

plt.figure(1)

plt.subplot(2,2,1)
plt.title('linear')
plt.plot(x,x)

#equivalent to plt.subplot(2,2,2)
plt.subplot(2,2,2)
plt.title('cubic')
plt.plot(x,x**3)

plt.figure(2)

plt.subplot(2,2,1)
plt.title('log')
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title('exponential')
plt.plot(x,x**2)
plt.show()
```

# pie chart

```python
import matplotlib.pyplot as plt
slices=[59267,56567,35566,78877,68787]
labels=['python','html','sql','java','css']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'},startangle=90,shadow=True)

plt.style.use('fivethirtyeight')

plt.title('pie chart')
plt.tight_layout()

plt.grid()
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

# stack plot

```python
import matplotlib.pyplot as plt
minutes=[1,2,3,4,5,6]
player1=[8,6,5,5,4,3]
player2=[6,1,7,5,2,9]
player3=[3,6,6,0,4,3]
labels=['player1','player2','player3']
colors=['#6d904f','#fc4f30','#008fd5']
plt.stackplot(minutes,player1,player2,player3,labels=labels,colors=colors)

plt.style.use('fivethirtyeight')

plt.title('stack plot')
plt.tight_layout()

plt.grid()
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

# fill_between

```python
import matplotlib.pyplot as plt
xaxis=[25,26,27,28,29,30]
y1_axis=[28000,29000,30000,31000,32000,33000]
plt.plot(xaxis,y1_axis,color='k',linestyle='--',marker='o',linewidth=2,label='y1_axis')
y2_axis=[28000,32000,49000,50000,65000,75000]
plt.plot(xaxis,y2_axis,color='#5a7d9a',marker='d',label='y2_axis')
plt.fill_between(xaxis,y1_axis,y2_axis,color='red',alpha=0.25)
plt.style.use('fivethirtyeight')
plt.xlabel('ages')
plt.ylabel('salary')
plt.title('fill_between')
plt.tight_layout()
plt.legend()
plt.grid(False)
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

# histogram

```python
import matplotlib.pyplot as plt
xaxis=[14,19,28,22,30,28]
bins=[10,20,30,40,50,60]
plt.hist(xaxis,bins,edgecolor='black')

age=29
color='#fc4f30'
plt.axvline(age,color=color,label='age',linewidth=2)
plt.style.use('fivethirtyeight')

plt.title('histogram plot')
plt.tight_layout()
plt.legend()
plt.grid(False)
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

# scatter plot

```python
import matplotlib.pyplot as plt
xaxis=[14,16,14,18,17,20,19]
bins=[10,20,30,40,50,60,70]
plt.scatter(xaxis,bins,label='scatter')
cbar=plt.colorbar()
cbar.set_label('like')

plt.style.use('fivethirtyeight')

plt.title('scatter plot')
plt.tight_layout()
plt.legend()
plt.grid(False)
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

# time series

```python
import matplotlib.pyplot as plt
xaxis=[14,16,14,18,17,20,19]
bins=[10,20,30,40,50,60,70]
plt.plot_date(xaxis,bins,label='timeseries')

plt.gcf().autofmt_xdate()
plt.style.use('fivethirtyeight')

plt.title('time series plot')
plt.tight_layout()
plt.legend()
plt.grid(False)
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```
